```python
directory = '/data/stemness/br/nguyen/fpkm/'
```

```python
sample_fp_tups = [(fp, os.path.join(directory, fp))
       for fp in os.listdir(directory)]
```

```python
df = pd.read_csv(sample_fp_tups[0][1], sep=r' +|\t')
df
```

```python

df = pd.read_csv(sample_fp_tups[0][1], sep=r' +|\t')
genes = np.asarray(df['gene_id'])

expression_df = None
for i, (sample, fp) in enumerate(sample_fp_tups):
    if i % 50 == 0:
        print(i, sample)
    try:
        df = pd.read_csv(fp, sep=r' ')
        df = df.set_index('gene_id')
        df = df.loc[genes]
        df.columns = [sample.split('.')[0]]

        if expression_df is None:
            expression_df = df.copy()
        else:
            expression_df = pd.concat((expression_df, df), axis=1)
    except KeyError:
        print(f'failed: {sample}')

expression_df
```

```python
mg = mygene.MyGeneInfo()
```

```python
result = mg.querymany(list(expression_df.index), scopes='ensemblgene', fields='symbol', species='human')
```

```python
result[0]
```

```python
gene_id_to_symbol = {d['query']:d['symbol'] for d in result if 'symbol' in d}
```

```python
mask = np.asarray([True if g in gene_id_to_symbol else False for g in expression_df.index])
filtered = expression_df[mask]
filtered.index = [gene_id_to_symbol[x] for x in filtered.index]
filtered.index.name = 'gene'
filtered = filtered.groupby('gene').mean()
filtered
```

```python
filtered.to_csv('/data/stemness/br/nguyen/combined.tsv', sep='\t')
```
